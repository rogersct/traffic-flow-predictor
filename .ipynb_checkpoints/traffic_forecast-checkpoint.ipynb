```python
# Detect vehicles in image
from PIL import Image
import cv2
import numpy as np
import requests

# Read Image
image = Image.open("./dataset/2022-02-11 12-20-03.jpg")
image_arr = np.array(image)

# Convert Image to Grayscale
grey = cv2.cvtColor(image_arr,cv2.COLOR_BGR2GRAY)
Image.fromarray(grey)

# Apply gaussian blur (to remove noise)
blur = cv2.GaussianBlur(grey,(3,3),0)
Image.fromarray(blur)

# Apply morpholohical techniques - dilate
dilated = cv2.dilate(blur,np.ones((3,3)))
Image.fromarray(dilated)

# Apply morpholohical transformation with a kernel
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (2, 2))
closing = cv2.morphologyEx(dilated, cv2.MORPH_CLOSE, kernel) 
Image.fromarray(closing)

# Using car cascade to detect vehicle
car_cascade_src = 'haarcascade_car.xml'
car_cascade = cv2.CascadeClassifier(car_cascade_src)
cars = car_cascade.detectMultiScale(closing, 1.1, 1)
cars

cnt = 0
for (x,y,w,h) in cars:
    cv2.rectangle(image_arr,(x,y),(x+w,y+h),(255,0,0),2)
    cnt += 1
print(cnt, " cars found")
Image.fromarray(image_arr)

# We see that the detection of vehicles is bad. As such, we'll need a better vehicle detection model. 
```

```python
import os
import pandas as pd
import cv2
import csv
import collections
import numpy as np
from tracker import *

# Initialize Tracker
tracker = EuclideanDistTracker()
# Detection confidence threshold
confThreshold = 0.2
nmsThreshold= 0.2

# Middle cross line position
middle_line_position = 225   
up_line_position = middle_line_position - 15
down_line_position = middle_line_position + 15

# Store Coco Names in a list
classesFile = "coco.names"
classNames = open(classesFile).read().strip().split('\n')
print(classNames)
print(len(classNames))

## Model Files
modelConfiguration = 'yolov3-320.cfg'
modelWeights = 'yolov3-320.weights'
# configure the network model
net = cv2.dnn.readNetFromDarknet(modelConfiguration, modelWeights)
# Configure the network backend
net.setPreferableBackend(cv2.dnn.DNN_BACKEND_CUDA)
net.setPreferableTarget(cv2.dnn.DNN_TARGET_CUDA)
# Define random colour for each class
np.random.seed(42)
colors = np.random.randint(0, 255, size=(len(classNames), 3), dtype='uint8')

#img = cv2.imread("./dataset/2022-02-11 13-56-46.jpg")
#cv2.imshow('img', img)
#cv2.waitKey(0)

def postProcess(outputs,img):
    required_class_index = [2, 3, 5, 7]
    detected_classNames = []
    #global detected_classNames
    height, width = img.shape[:2]
    boxes = []
    classIds = []
    confidence_scores = []
    detection = []
    for output in outputs:
        for det in output:
            scores = det[5:]
            classId = np.argmax(scores)
            confidence = scores[classId]
            if classId in required_class_index:
                if confidence > confThreshold:
                    # print(classId)
                    w,h = int(det[2]*width) , int(det[3]*height)
                    x,y = int((det[0]*width)-w/2) , int((det[1]*height)-h/2)
                    boxes.append([x,y,w,h])
                    classIds.append(classId)
                    confidence_scores.append(float(confidence))
                    
    # Apply Non-Max Suppression
    indices = cv2.dnn.NMSBoxes(boxes, confidence_scores, confThreshold, 
nmsThreshold)
    # print(classIds)
    if len(indices)>0:
        for i in indices.flatten():
            x, y, w, h = boxes[i][0], boxes[i][1], boxes[i][2], boxes[i][3]
            # print(x,y,w,h)
            color = [int(c) for c in colors[classIds[i]]]
            name = classNames[classIds[i]]
            detected_classNames.append(name)
            # Draw classname and confidence score 
            cv2.putText(img,f'{name.upper()} {int(confidence_scores[i]*100)}%',
                      (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 1)
            # Draw bounding rectangle
            cv2.rectangle(img, (x, y), (x + w, y + h), color, 1)
            detection.append([x, y, w, h, required_class_index.index(classIds[i])])
            
    return detected_classNames

def from_static_image(image):
    input_size = 320
    img = cv2.imread(image)
    blob = cv2.dnn.blobFromImage(img, 1 / 255, (input_size, input_size), [0, 0, 0], 1, 
                   crop=False)
    # Set the input of the network
    net.setInput(blob)
    layersNames = net.getLayerNames()
    outputNames = [(layersNames[i - 1]) for i in net.getUnconnectedOutLayers()]
    # Feed data to the network
    outputs = net.forward(outputNames)
    # Find the objects from the network output
    detected_classNames = postProcess(outputs,img)
    # count the frequency of detected classes
    frequency = collections.Counter(detected_classNames)
    #print(frequency)
    
    num_vehicles = 0
    for freq_item in frequency:
        num_vehicles  = num_vehicles + frequency[freq_item]
    
    font_size = 0.5
    font_color =  (0, 0, 255)
    font_thickness = 2
    
       # Draw counting texts in the frame
    cv2.putText(img, "Car:        "+str(frequency['car']), (20, 40),
     cv2.FONT_HERSHEY_SIMPLEX, font_size, font_color, font_thickness)
    cv2.putText(img, "Motorbike:  "+str(frequency['motorbike']), (20, 60), 
cv2.FONT_HERSHEY_SIMPLEX, font_size, font_color, font_thickness)
    cv2.putText(img, "Bus:        "+str(frequency['bus']), (20, 80), 
cv2.FONT_HERSHEY_SIMPLEX, font_size, font_color, font_thickness)
    cv2.putText(img, "Truck:      "+str(frequency['truck']), (20, 100), 
cv2.FONT_HERSHEY_SIMPLEX, font_size, font_color, font_thickness)
    cv2.imshow("image", img)
    cv2.imwrite("./result.jpg", img)
    cv2.waitKey(0)
#     with open("static-data.csv", 'a') as f1:
#         cwriter = csv.writer(f1)
#         cwriter.writerow([image, frequency['car'], frequency['motorbike'], frequency['bus'], 
# frequency['truck']])
#         f1.close()

    return num_vehicles

if __name__ == '__main__':
    # realTime()
    #image_file = "./dataset/2022-02-11 13-56-46.jpg"
    image_file = "./dataset/2022-02-11 12-20-03.jpg" # Truck
    #image_file = "./dataset/2022-02-11 16-19-51.jpg" # Bus        
    from_static_image(image_file)
    
#     data_folder = "./dataset/"
    
#     # Initialize empty dataframe with columns names
#     df = pd.DataFrame(columns=['Timestamp','Vehicle_Count'])
    
#     # Iterate over files in datafolder
#     for root, dirs, files in os.walk(data_folder):
#         for filename in files:
#             image_file = os.path.join(root, filename)
#             # Perform vehicle detection
#             num_vehicles = from_static_image(image_file)
#             timestamp = os.path.splitext(filename)[0]
#             print(timestamp, num_vehicles)
#             # Populate row data
#             new_row = {'Timestamp': timestamp, 'Vehicle_Count':num_vehicles}
#             #Append row to the dataframe
#             df = df.append(new_row, ignore_index=True)
            

#     # Write dataframe to file for further analysis
#     df.to_csv('./traffic_count.csv', index=False)  
```

Through the chart, we can see a clear pattern whereby number of vehicles on the specific lat and long location get lowest(0) at 12am -6am and the traffic starts to pick up after 6am. However, as we see, the chart is quite noisy. Hence we'll do some resampling to smoother out the curves.

```python
df_traffic_count_index = df_traffic_count.set_index(['Timestamp'])
df_traffic_count_index = df_traffic_count_index.resample('60T', label='right', closed='right').sum()
df_traffic_count_index_org = df_traffic_count_index.copy()

fig = plt.figure(figsize=(12, 6))
# plotting the points
plt.plot(df_traffic_count_index.index.values, df_traffic_count_index['Vehicle_Count'].values)
plt.xlabel('Time')
plt.ylabel('Number of Vehicles')
plt.grid(True)
plt.title('Resampled 1 Hour Interval, Number of Vehicles Detected')
plt.show()
```

```python
%matplotlib inline
from statsmodels.tsa.stattools import adfuller
# Test for stationary
test_result=adfuller(df_traffic_count_index['Vehicle_Count'])

# Using null hypothesis, H0 (null hypothesis) statement about the population is true, H1 (alternative hypothesis) claims
# that the population is contradictory to H0. H0 is non-stationary, H1 is stationary.

def adfuller_test(sales):
    result=adfuller(sales)
    labels = ['ADF Test Statistic','p-value','#Lags Used','Number of Observations']
    for value,label in zip(result,labels):
        print(label+' : '+str(value) )

    if result[1] <= 0.05:
        print("Strong evidence against the null hypothesis(Ho), reject the null hypothesis. Data is stationary")
    else:
        print("Weak evidence against null hypothesis,indicating it is non-stationary ")

adfuller_test(df_traffic_count_index['Vehicle_Count'])

from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
import statsmodels.api as sm
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(df_traffic_count_index['Vehicle_Count'].dropna(),lags=10,ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(df_traffic_count_index['Vehicle_Count'].dropna(),lags=10,ax=ax2)
```

ACF explains how present value of a given time series is correlated with the past 1, 2, n values. 
PACF is the partial autocorrelation fucntion between the series and lags of itself. 
Plotting ACF and PACF to identify the orders of AR and MA terms. 
To determine AR process, ACF plot will gradually decrease whereas PACF will have a sharp drop after p significant lags. 
ARMA(1,1)

```python
import numpy as np
from statsmodels.tsa.stattools import acf

# Create Training and Test
# 85% Train , 15% Test
train_size = int(0.85 * df_traffic_count_index.shape[0])
x_train = df_traffic_count_index['Vehicle_Count'][:train_size]
x_test = df_traffic_count_index['Vehicle_Count'][train_size:]
print(train_size)

# Build Arima Model with training data
model=sm.tsa.statespace.SARIMAX(x_train ,order=(1, 0, 1),seasonal_order=(1,1,0,24))
#model=sm.tsa.statespace.SARIMAX(x_train ,order=(1, 1, 0),seasonal_order=(2,1,0,24))
results=model.fit()

fc = results.predict(start = train_size, end = df_traffic_count_index.shape[0]-1, dynamic= True)

# Make as pandas series
fc_series = pd.Series(fc, index=x_test.index)

# Plot
plt.figure(figsize=(12,6))
plt.plot(x_train, label='training')
plt.plot(x_test, label='actual')
plt.plot(fc_series, label='forecast')
plt.xlabel('Time')
plt.ylabel('Number of Vehicles per Hour')
plt.title('Forecast vs Actuals')
plt.grid(True)
plt.legend(loc='upper right', fontsize=8)
plt.show()

# Accuracy metrics
def forecast_accuracy(forecast, actual):
    
    mape = np.mean(np.abs(forecast - actual)/np.abs(actual))  # MAPE

    me = np.mean(forecast - actual)             # ME
    mae = np.mean(np.abs(forecast - actual))    # MAE
    mpe = np.mean((forecast - actual)/actual)   # MPE
    rmse = np.mean((forecast - actual)**2)**.5  # RMSE
    corr = np.corrcoef(forecast, actual)[0,1]   # corr
    mins = np.amin(np.hstack([forecast[:,None], 
                              actual[:,None]]), axis=1)
    maxs = np.amax(np.hstack([forecast[:,None], 
                              actual[:,None]]), axis=1)
    #minmax = 1 - np.mean(mins/maxs)             # minmax
    #acf1 = acf(fc-x_test)[1]                      # ACF1
    r2score = round(metrics.r2_score(actual, forecast), 4)
    return({'mape':mape, 'me':me, 'mae': mae, 
            'mpe': mpe, 'rmse':rmse, 'corr':corr, 'r^2 score': r2score})

forecast_accuracy(fc.values, x_test.values)
```

Typically, the MAPE, Correlation and Min-Max Error can be used. As the three are percentage errors that vary between 0 and That way, you can judge how good is the forecast irrespective of the scale of the series.

The other error metrics are quantities. For example, an RMSE of 100 for a series whose mean is in 1000s is better than an RMSE of 5 for series in 10s. So, you cant really use them to compare the forecasts of two different scaled time series.

```python
# Using the automatic SARIMA
import pmdarima as pm

# Seasonal - fit stepwise auto-ARIMA
smodel = pm.auto_arima(x_train, start_p=1, start_q=1,
                         test='adf',
                         max_p=3, max_q=3, m=24,
                         start_P=0, seasonal=True,
                         d=None, D=1, trace=True,
                         error_action='ignore',  
                         suppress_warnings=True, 
                         stepwise=True)

smodel.summary()
```

```python
# 3 Hour Forecast
n_periods = 51 
fitted, confint = smodel.predict(n_periods=n_periods, return_conf_int=True)
index_of_fc = pd.date_range(df_traffic_count_index.index[-1], periods = n_periods, freq='1H')

# Series for plotting purpose
fitted_series = pd.Series(fitted, index=index_of_fc)
lower_series = pd.Series(confint[:, 0], index=index_of_fc)
upper_series = pd.Series(confint[:, 1], index=index_of_fc)

# Plot
plt.plot(df_traffic_count_index['Vehicle_Count'])
plt.plot(fitted_series, color='darkgreen')
plt.grid(True)
plt.fill_between(lower_series.index, 
                 lower_series, 
                 upper_series, 
                 color='k', alpha=.15)

plt.title("SARIMA - Forecast")
plt.show()

forecast_accuracy(fitted, x_test.values)
```

We see that by using the stepwise approach to search through multiple combinations of p,d,q paramters. The best model is chosen based on the least AIC.

Akaike Information Critera (AIC) is used to measure the goodness of fit for a statiscal model.

```python
from statsmodels.tsa.api import VAR
import statsmodels.api as sm

feature_names = ['Vehicle_Count', 'Rainfall']
X = df_merged[feature_names]

# 85% Train , 15% Test
train_size = int(0.85 * df_merged.shape[0])
x_train = X[:train_size]
x_test = X[train_size:]

model  = sm.tsa.VARMAX(x_train, order=(1,0))
# Fit the model via maximum likelihood, using the EM algorithm
model_result = model.fit()
# Note that res_dfm is an instance of the DynamicFactorMQResults class

# Show the summary of results
print(model_result.summary())

# To determine the lag order
for i in range(5):
    i += 1
    model = sm.tsa.VARMAX(x_train, order=(i,0))
    model_result = model.fit()
    print('Order = ', i)
    print('AIC: ', model_result.aic)
    print('BIC: ', model_result.bic)
    print('HQIC: ', model_result.hqic)

# Initialize Multivarite ARMAX model
model = sm.tsa.VARMAX(x_train, order=(2, 0), trend='c')
model_result = model.fit()

# Forecast the next number of steps
fc = model_result.forecast(steps=len(x_test))
print(fc.shape)

# Make as pandas series
fc_series = pd.Series(fc['Vehicle_Count'], index=x_test.index)

# Plot forecast vs actual
plt.figure(figsize=(12,6))
plt.plot(x_train['Vehicle_Count'], label='training')
plt.plot(x_test['Vehicle_Count'], label='actual')
plt.plot(fc_series, label='forecast')
plt.title('Forecast vs Actuals')
plt.grid(True)
plt.legend(loc='upper right', fontsize=8)
plt.show()

# Model Evaluation
forecast_accuracy(fc['Vehicle_Count'].values, x_test['Vehicle_Count'].values)

```

As shown in the graph above, the VARMAX is unable to forecast the vehicle count accurately. No seasonality data is being learnt in VARMAX. As such, we can see the performance metrics like mae, rmse has increased and a very low r^2 score.

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import ParameterGrid
from sklearn import metrics

# Extract the seasonality data
df_merged = pd.merge(df_traffic_count_index_org, df_rainfall_index,
                        how="left", on=["Timestamp"])
df_merged.reset_index(inplace=True)
df_merged['Year'] = pd.DatetimeIndex(df_merged['Timestamp']).year
df_merged['Month'] = pd.DatetimeIndex(df_merged['Timestamp']).month
df_merged['Day'] = pd.DatetimeIndex(df_merged['Timestamp']).day
df_merged['Hour'] = pd.DatetimeIndex(df_merged['Timestamp']).hour

feature_names = ['Year','Month','Day','Hour','Rainfall']
X = df_merged[feature_names]
y = df_merged['Vehicle_Count']

# 85% Train , 15% Test
train_size = int(0.85 * df_merged.shape[0])
X_train = X[:train_size]
y_train = y[:train_size]
X_test = X[train_size:]
y_test = y[train_size:]
y_test_time = df_merged['Timestamp'][train_size:]

# Scale - Standardization is known to improve the numerical stability of the model
scaler = MinMaxScaler(feature_range=(0,1))
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Using ParameterGrid to identify the best parameter set for training the random forest model
grid = {'n_estimators': [1000], 'max_features': [1, len(feature_names)], 'random_state': [42]}
test_scores = []

rf_model = RandomForestRegressor()

for g in ParameterGrid(grid):
    rf_model.set_params(**g) 
    rf_model.fit(X_train, y_train)
    test_scores.append(rf_model.score(X_test, y_test))

best_index = np.argmax(test_scores)
print(test_scores[best_index], ParameterGrid(grid)[best_index])

# Initialize Random Forest Model
rf_model = RandomForestRegressor(n_estimators=1000, max_features=5, random_state=42)
rf_model.fit(X_train, y_train)

y_pred = rf_model.predict(X_test)

# Plot to visualize the predicted price change percentage over the time.
fig = plt.figure()
y_pred_series = pd.Series(y_pred, index=y_test.index)
#y_pred_series.plot()
plt.plot(y_test_time, y_pred, label='forecast')
plt.plot(y_test_time, y_test, label='actual')
plt.grid(True)
plt.xlabel('Time')
plt.ylabel('Number of Vehicles per Hour')
plt.title('Forecast vs Actuals')
plt.grid(True)
plt.legend(loc='upper right', fontsize=8)
plt.show()

# # Model Evaluation
forecast_accuracy(y_pred, y_test)
```

As compared to the univariate SARIMAX model, the performance of the random forest model has lower mae, rmse and higher corr and r^2 score. The higher the r^2 score the better model fits the observed data. 

This shows that using random forest to forecast traffic flow using multiple variables yields better performance as compared to SARIMAX model. 

However, speaking of which, this model is far from ideal. There are many other variables (depending on the availablity of the data) that we can include such as speed of the vehicles on the current road segment, speed of vehicles on "connected" road segments. We can use spatial radius to define the connectivity of certain road segments. 

Likewise, we can explore other types of models like Long Short Term Memory (a variant of Recurrent Neural Network) to capture the time series characteristics of traffic flow data
